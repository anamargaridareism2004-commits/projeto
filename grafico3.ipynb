```python
import matplotlib.pyplot as plt
import numpy as np
import clinic as c


def extraiFilaMedia(t):
    res = []
    for _, fila_media in t:
        res.append(fila_media)
    return res


def grafFilaVsLambda(repeticoes=20):

    resultados = []

    taxa_original = c.TAXA_CHEGADA

    for lam in range(10, 31):
        c.TAXA_CHEGADA = lam / 60.0

        filas = []
        for _ in range(repeticoes):
            res = c.simula()
            filas.append(res["tamanho_medio_fila"])

        resultados.append((lam, np.mean(filas)))

    c.TAXA_CHEGADA = taxa_original

    x = list(range(10, 31))
    y = extraiFilaMedia(resultados)

    plt.plot(x, y, label="Tamanho médio da fila")
    plt.title("Relação entre λ e o tamanho médio da fila")
    plt.xlabel("Taxa de chegada λ (doentes/hora)")
    plt.ylabel("Tamanho médio da fila")
    plt.legend()
    plt.show()


if __name__ == "__main__":
    grafFilaVsLambda(repeticoes=20)
```
